```python
class Person:
    def __init__(self, name):
        self.name = name
        
    def calculate_long_stuff(self):
        if not hasattr(self, "_cached_stuff"):
            self._cached_stuff = sum([x for x in range(10000000)])
        return self._cached_stuff
```

```python
a = Person("Adam")
a.calculate_long_stuff()
```

```python
class Boy(Person):
    def __init__(self, name, age):
        super().__init__(name)
        self.arg1 = 10
        self.arg2 = 20
        self.arg3 = 100
```

```python
boy1 = Boy("Adam", 10)
```

```python
args_to_fetch = ["arg1", "arg2", "arg3"]
for i in args_to_fetch:
    print(getattr(boy1, i))
```

```python
class Girl(Person):
    if_class_variable = True
    
    def __init__(self, name):
        super().__init__(name)
        self.if_class_variable = False
    
    @classmethod
    def get_class_variable(cls):
        print(cls.if_class_variable)
        
```

```python
Girl.if_class_variable
```

```python
girl1 = Girl("Jenny")
girl1.if_class_variable
```

```python
from collections import namedtuple
from math import log

OneCalculation = namedtuple("OneCalculation", "type x y result")

class Calc:
    def __init__(self, name = "Calc", producer = "Casio", color = "Black", mem_list = []):
        self.name = name
        self.producer = producer
        self.color = color
        self.mem_list = mem_list
        
    
    # Add,Substract,Divide,Multiply,PrintResults
    def add(self, x, y):
        self.mem_list.append(OneCalculation(type = "add", x = x, y = y, result = x+y))
        
        print(self.mem_list[-1].result)
    
    def substract(self, x, y):
        self.mem_list.append(OneCalculation(type = "substract", x = x, y = y, result = x-y))
        
        print(self.mem_list[-1].result)
        
    def divide(self, x, y):
        try:
            self.mem_list.append(OneCalculation(type = "divide", x = x, y = y, result = x/y))
            print(self.mem_list[-1].result)
            
        except ZeroDivisionError:
            print("Dividing by 0. Results not saved")
        
        
        
    def multiply(self, x, y):
        self.mem_list.append(OneCalculation(type = "multiply", x = x, y = y, result = x*y))
        
        print(self.mem_list[-1].result)
        
    def print_results(self):
        for i in self.mem_list:
            print(f"Type of calculation: {i.type}, x = {i.x}, y = {i.y}, result: {i.result} ")
    
    
        
```

```python
class ScientificCalc(Calc):
    def __init__(self, name = "Calc", producer = "Casio", color = "Black", mem_list = [], version = "2.1"):
        super().__init__(name, producer, color, mem_list)
        
    def pow(self, x, y):
        self.mem_list.append(OneCalculation(type = "power", x = x, y = y, result = x**y))
        
        print(self.mem_list[-1].result)
    
    def log(self, x):
        self.mem_list.append(OneCalculation(type = "log", x = x, y = "", result = log(x)))
        
        print(self.mem_list[-1].result)
        
```

```python
a = Calc()
a.add(1,2)
a.substract(3,4)
a.print_results()
```
